```python
import PIL
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt

```

```python
image_folder='./clue_scroll_rewards/'
images = []
for dirpath, dnames, fnames in os.walk(image_folder):
    for f in fnames:
        images.append(f)
positive = 'Medium(5286) 2024-03-07_21-17-58.png'
negative = 'Medium(3047) 2023-08-10_09-05-08.png'
#images = [negative, positive]
ranger = './rangers.png'
#todo crop clue to loot area before processing
image_rgb = cv2.imread(image_folder+positive)
y = 40
x = 140
h = 250
w = 250
crop_img = image_rgb[y:y+h, x:x+w]
plt.imshow(crop_img)
```

```python
for image in images:
    img_rgb = cv2.imread(image_folder+image)
    template = cv2.imread(ranger)
    w, h = template.shape[:-1]
    res = cv2.matchTemplate(img_rgb, template, cv2.TM_CCOEFF_NORMED)
    threshold = .8
    loc = np.where(res >= threshold)
    if loc:
        print(image)
        for pt in zip(*loc[::-1]):  # Switch columns and rows
            cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0, 0, 255), 2)
            cv2.imwrite(image, img_rgb)
```
